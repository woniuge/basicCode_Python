```python
import numpy as np
import pandas as pd
```

```python
# 评分
rnames = ['user_id','movie_id','rating','timestamp']
rpath = 'ratings.csv'
ratings = pd.read_csv(filepath_or_buffer=rpath,sep=',',header=0,names=rnames,engine='python')
ratings.head()
```

```python
# 电影信息
mnames = ['movie_id','title','genres']
mpath = 'movies.csv'
movies = pd.read_csv(filepath_or_buffer=mpath,sep=',',header=0,names=mnames,engine='python')
movies.head()
```

```python
# 链接信息
lnames = ['movie_id','imdbId','tmdbId']
lpath = 'links.csv'
links = pd.read_csv(filepath_or_buffer=lpath,sep=',',header=0,names=lnames,engine='python')
links.head()
```

```python
# 标签信息
tnames = ['user_id','movie_id','tag','timestamp']
tpath = 'tags.csv'
tags = pd.read_csv(filepath_or_buffer=tpath,sep=',',header=0,names=tnames,engine='python')
tags.head()
```

```python
# 合并数据集
data = pd.merge(ratings,movies,on='movie_id')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# 汇总每部电影的评分数量
rating_count_by_movie = data.groupby(['movie_id','title'],as_index=False)['rating'].count()
rating_count_by_movie
```

```python
rating_count_by_movie.columns = ['movie_id','title','rating_count']
rating_count_by_movie
```

```python
rating_count_by_movie.sort_values(by=['rating_count'],ascending=False,inplace=True)
rating_count_by_movie
```

```python
rating_count_by_movie.info()
```

```python
# 得到打分的平均值及方差
rating_stddev = data.groupby(['movie_id','title']).agg({'rating':['mean','std']})
rating_stddev
```
